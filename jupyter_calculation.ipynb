```python
# change data type of price columns to make further calculations (if there is ',' change it to '.' first)

all_countries['product_regular_price'] = all_countries['product_regular_price'].astype(float)
all_countries['product_regular_price_pln'] = all_countries['product_regular_price_pln'].astype(float)

# to change type to float in column there has to be some value - not Nan
all_countries['product_promo_price'] = all_countries['product_promo_price'].fillna('0')
all_countries['product_promo_price'] = all_countries['product_promo_price'].astype(float)
all_countries['product_promo_price_pln'] = all_countries['product_promo_price_pln'].fillna('0')
all_countries['product_promo_price_pln'] = all_countries['product_promo_price_pln'].astype(float)
```

```python
# create two new columns and sign them value 0

all_countries['regular_price_diff_to_poland'] = 0
all_countries['promo_price_diff_to_poland'] = 0

all_countries.head()
```

```python
# calculate differences compare to Poland price (PLN) for new columns

select_poland = (all_countries['country'] == 'Poland')
not_poland_countries = ['Russia', 'Singapore', 'Thailand', 'United States', 'Zimbabwe']
select_not_poland = (all_countries['country'].isin(not_poland_countries))

# has to be float data type

all_countries['regular_price_diff_to_poland'] = all_countries['regular_price_diff_to_poland'].astype(float)
all_countries['promo_price_diff_to_poland'] = all_countries['promo_price_diff_to_poland'].astype(float)
```

```python
for index,row in all_countries.loc[select_not_poland].iterrows():
    for index_poland,row_poland in poland.iterrows():
        if row['product_id'] == row_poland['product_id']:
            country = row['country']
            regular_price_poland = row_poland['product_regular_price_pln']
            promo_price_poland = row_poland['product_promo_price_pln']
            regular_price_country = row['product_regular_price_pln']
            promo_price_country = row['product_promo_price_pln']
            all_countries.at[row['id'], 'regular_price_diff_to_poland'] = (regular_price_country - regular_price_poland)
            all_countries.at[row['id'], 'promo_price_diff_to_poland'] = (promo_price_country - promo_price_poland)

all_countries.to_csv('all_countries.csv', sep=';', encoding='utf-8', index=False)
```

```python
# round regular_price_diff_to_poland, promo_price_diff_to_poland to 2 decimal places

all_countries = pd.read_csv('all_countries.csv', header=0, delimiter=';')

cols = ['regular_price_diff_to_poland', 'promo_price_diff_to_poland']
all_countries[cols] = all_countries[cols].round(2)

all_countries.to_csv('all_countries.csv', sep=';', encoding='utf-8', index=False)
```

```python
# ADD ANOTHER COLUMN - DIFFERENCE BETWEEN REGULAR PRICE AND PROMO PRICE
all_countries = pd.read_csv('all_countries.csv', header=0, delimiter=';')

all_countries['diff_regular_to_promo_price'] = 0
all_countries.head()
```

```python
# has to be float data type
all_countries['diff_regular_to_promo_price'] = all_countries['diff_regular_to_promo_price'].astype(float)
```

```python
# calculate differences between reguar price and promo price
select_sale = (all_countries['master_category'] == 'SALE')
for index,row in all_countries.loc[select_sale].iterrows():
    regular_price_pln = row['product_regular_price_pln']
    promo_price_pln = row['product_promo_price_pln']
    all_countries.at[row['id'], 'diff_regular_to_promo_price'] = (regular_price_pln - promo_price_pln)
```

```python
# round diff_regular_to_promo_priced to 2 decimal places
all_countries['diff_regular_to_promo_price'] = all_countries['diff_regular_to_promo_price'].round(2)
```
